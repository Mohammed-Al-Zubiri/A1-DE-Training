# Task 2: Built-in Packages and Modules in Python

Python comes with a lot of useful tools already installed - these are called **built-in modules** (or the **standard library**). We don't need to install anything extra to use them, we just import them.

## What's the difference between a Module and a Package?

- A **module** is just a single Python file that contains functions, classes, etc, like `math.py`
- A **package** is a folder that contains multiple modules, like `os`

## 1. math module

The `math` module gives us mathematical functions like square root, power, pi, etc.

```python
import math

# square root
print(f"Square root of 16: {math.sqrt(16)}")

# power
print(f"2 to the power 3: {math.pow(2, 3)}")

# pi value
print(f"Value of pi: {math.pi}")

# factorial
print(f"Factorial of 5: {math.factorial(5)}")

# ceiling and floor
print(f"Ceiling of 4.3: {math.ceil(4.3)}")
print(f"Floor of 4.7: {math.floor(4.7)}")
```

## 2. random module

The `random` module helps us generate random numbers, pick random items, etc.

```python
import random

# random integer between 1 and 10
print(f"Random number between 1 and 10: {random.randint(1, 10)}")

# random float between 0 and 1
print(f"Random decimal: {round(random.random(), 4)}")

# pick a random item from a list
fruits = ["apple", "banana", "cherry", "mango"]
print(f"Random choice from list: {random.choice(fruits)}")

# shuffle a list
numbers = [1, 2, 3, 4, 5]
random.shuffle(numbers)
print(f"Shuffled list: {numbers}")
```

## 3. datetime module

This module is for working with dates and times. Very useful when we need to track when something happened.

```python
from datetime import datetime, date, timedelta

# current date and time
now = datetime.now()
print(f"Current date and time: {now.strftime('%Y-%m-%d %H:%M:%S')}")

# just today's date
today = date.today()
print(f"Just the date: {today}")

# create a specific date
my_date = date(2025, 1, 15)
print(f"Custom date: {my_date}")

# format date
print(f"Formatted: {today.strftime('%d/%m/%Y')}")

# difference between dates
new_year = date(2027, 1, 1)
diff = new_year - today
print(f"Days until new year: {diff.days} days")
```

## 4. os module

The `os` module lets us interact with the operating system - like checking files, folders, getting the current directory, etc.

```python
import os

# get current working directory
print(f"Current directory: {os.getcwd()}")

# list files in a directory
files = os.listdir('.')  # '.' means current directory
print(f"Files in current directory: {files[:3]}")  # show first 3

# check if a file exists
print(f"Does 'test.txt' exist? {os.path.exists('test.txt')}")

# get the operating system name
print(f"Operating system: {os.name}")
```

## 5. json module

The `json` module helps us work with JSON data (which is very common in web development and APIs).

```python
import json

# convert python dict to JSON string
student = {
    "name": "Ahmed",
    "age": 21,
    "courses": ["Python", "Data Science"]
}

json_string = json.dumps(student, indent=2)
print(f"Python dict to JSON string:\n{json_string}")

# convert JSON string back to python dict
json_data = '{"name": "Sara", "age": 22, "city": "Sana\'a"}'
parsed = json.loads(json_data)
print(f"\nJSON string back to Python dict:")
print(f"Name: {parsed['name']}, City: {parsed['city']}")
```

## 6. sys module

The `sys` module gives us access to system-specific things like the Python version, command line arguments, etc.

```python
import sys

# python version
print(f"Python version: {sys.version.split()[0]}")

# platform
print(f"Platform: {sys.platform}")

# max integer size
print(f"Max integer size: {sys.maxsize}")
```

## 7. string module

The `string` module has useful string constants and a Template class for simple text formatting.

```python
import string

# useful constants
print(f"All ASCII letters: {string.ascii_letters}")
print(f"Digits: {string.digits}")
print(f"Punctuation: {string.punctuation}")

# simple password generator using string + random
import random
chars = string.ascii_letters + string.digits + "!@#$"
password = ''.join(random.choice(chars) for i in range(8))
print(f"Random password: {password}")
```

## 8. collections module

This module provides special container types that are more powerful than regular lists and dicts.

```python
from collections import Counter, OrderedDict, defaultdict

# Counter - count how many times each item appears
words = ["hello", "world", "hello", "python", "hello", "world"]
word_count = Counter(words)
print(f"Word frequency: {word_count}")
print(f"Most common: {word_count.most_common(2)}")

# OrderedDict - remembers the order items were added
od = OrderedDict()
od['first'] = 1
od['second'] = 2
od['third'] = 3
print(f"\nOrderedDict keeps insertion order:")
for key, val in od.items():
    print(f"{key} -> {val}")

# defaultdict - dict with default values
fruit_count = defaultdict(int)  # default value is 0
fruits = ["apple", "banana", "apple", "cherry", "banana", "banana"]
for fruit in fruits:
    fruit_count[fruit] += 1
print(f"\ndefaultdict with default value 0:")
print(fruit_count)
```
