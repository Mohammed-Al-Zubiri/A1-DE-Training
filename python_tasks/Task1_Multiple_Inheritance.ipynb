# Task 1: Multiple Inheritance in Python

Multiple inheritance is when a class can inherit from **more than one parent class**. This means the child class gets the attributes and methods from all its parents.

## Multiple Inheritance - Example

In multiple inheritance, a child class inherits from **two or more** parent classes. Python supports this, unlike some other languages like Java.

````python
# multiple inheritance example
class Father:
    def __init__(self):
        self.last_name = "Smith"
    
    def get_last_name(self):
        return self.last_name

class Mother:
    def __init__(self):
        self.eye_color = "Blue"
    
    def get_eye_color(self):
        return self.eye_color

class Child(Father, Mother):  # inherits from both parents
    def __init__(self, name):
        Father.__init__(self)
        Mother.__init__(self)
        self.name = name

# testing it
father = Father()
mother = Mother()
child = Child("John")

print(f"Father's last name: {father.get_last_name()}")
print(f"Mother's eye color: {mother.get_eye_color()}")
print(f"Child's name: {child.name}")
print(f"Child got last name: {child.get_last_name()}")  # from Father
print(f"Child got eye color: {child.get_eye_color()}")  # from Mother
````

As we can see, the `Child` class got `get_last_name()` from `Father` and `get_eye_color()` from `Mother`. That's the idea of multiple inheritance - getting stuff from multiple parents.

## The Diamond Problem

One tricky thing with multiple inheritance is the **diamond problem**. This happens when two parent classes inherit from the same grandparent class.

It looks like a diamond shape:
```
      A
     / \
    B   C
     \ /
      D
```

The question is: if D calls a method that exists in A, B, and C, which one does it use?

````python
# diamond problem example
class A:
    def greet(self):
        print("Hello from A")

class B(A):
    def greet(self):
        print("Hello from B")

class C(A):
    def greet(self):
        print("Hello from C")

class D(B, C):  # inherits from both B and C
    pass

d = D()
d.greet()  # which greet() will it call?
````

It called `B`'s greet! But why? This is because of something called **MRO**.

## MRO (Method Resolution Order)

MRO is the order Python uses to look for methods when there's multiple inheritance. Python uses an algorithm called **C3 Linearization** to figure out this order.

Basically, Python checks the classes from left to right in the inheritance list, then goes up to the parents. We can see the MRO using the `mro()` method or `__mro__` attribute.

````python
# checking the MRO
print("MRO of D:")
print(" -> ".join([cls.__name__ for cls in D.mro()]))
````

So the order is: D → B → C → A → object

That's why `B.greet()` was called first - Python checks B before C because B comes first in `class D(B, C)`.

## Using super() with Multiple Inheritance

`super()` is used to call parent methods. With multiple inheritance, it follows the MRO to decide which parent to call.

````python
# using super() with multiple inheritance
class Animal:
    def __init__(self, name):
        self.name = name
        print(f"Animal created: {name}")

class Flyer(Animal):
    def __init__(self, name):
        super().__init__(name)
        print(f"Flying: {name}")
    
    def fly(self):
        return f"{self.name} is flying!"

class Swimmer(Animal):
    def __init__(self, name):
        super().__init__(name)
        print(f"Swimming: {name}")
    
    def swim(self):
        return f"{self.name} is swimming!"

class Eagle(Flyer, Swimmer):  # eagle can fly and swim
    def __init__(self, name):
        super().__init__(name)

eagle = Eagle("Eagle")
print(f"\n{eagle.fly().split(' is')[0]} can fly and swim!")
````

## Summary

- **Multiple Inheritance** lets a class inherit from more than one parent
- The **Diamond Problem** happens when two parents share the same grandparent
- Python solves this using **MRO** (Method Resolution Order)
- MRO follows the order we list the parents (left to right)
- `super()` follows the MRO chain automatically
