```python
import cosine_similarity as cs
```

```python
text1 = input("Enter the first text: ")
text2 = input("Enter the second text: ")
```

```python
print(text1)
print(text2)
```

```python
cleaned1 = cs.clean_text(text1)
cleaned2 = cs.clean_text(text2)
print(cleaned1)
print(cleaned2)
```

```python
bag_of_words = cs.generate_bag_of_words(cleaned1 + " " + cleaned2)
print(bag_of_words)
```

```python
vec1 = cs.generate_vector(bag_of_words, cleaned1)
vec2 = cs.generate_vector(bag_of_words, cleaned2)
print(vec1)
print(vec2)
```

```python
print("Cosine Similarity: ", cs.cosine_similarity(vec1, vec2))
```
